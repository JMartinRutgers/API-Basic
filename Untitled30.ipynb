```python
import pandas as pd
import io
import requests
import csv 
import pandas as pd
from scipy.special import ndtr as ndtr
import numpy as np
import scipy.stats as st
from sklearn.preprocessing import scale, normalize
import pickle
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style 

import seaborn as sns 
url="https://www.cbp.gov/sites/default/files/assets/documents/2022-Sep/nationwide-encounters-fy19-fy22-aug.csv"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
data= pd.DataFrame(c)
```

```python
data
```

```python
df = data.groupby(["Fiscal Year","Land Border Region"], as_index=False)['Citizenship','Demographic'].max()
df
```

```python
df = data.groupby(["Fiscal Year"], as_index=False)['Citizenship','Demographic'].max()
df 
```

```python
df = data.groupby(["Fiscal Year"])['Citizenship'].min()
df
```

```python
df = data.groupby(["Fiscal Year"])['Citizenship'].sum()
df
```

```python
df = data.groupby(["Citizenship"]).get_group("CUBA")
df
```

```python
df = data.groupby(["Citizenship"]).max()
# Create a bar chart based off of the group series from before
count_chart = df.plot(kind='barh')

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Fiscal Year")
count_chart.set_ylabel("Citizenship")


plt.show()
plt.tight_layout()
```
